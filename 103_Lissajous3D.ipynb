```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
#from mpl_toolkits.mplot3d import Axes3D

# パラメータ
w1, w2, w3 = 1, 2, 3
N = 1000
t = np.linspace(0, 2*np.pi, N)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
ax.set_title("3D Lissajous (Simple Version)")

ax.axis('off')
fig.patch.set_facecolor("black")
ax.set_facecolor("black")

# 空のline
line, = ax.plot([], [], [], lw=5)

def init():
    line.set_data([], [])
    line.set_3d_properties([])
    return line,

def update(frame):
    a = frame * 0.03 * np.pi

    x = np.sin(w1 * t)
    y = np.sin(w2 * t + a)
    z = np.sin(w3 * t + a*0.7)

    line.set_data(x, y)
    line.set_3d_properties(z)

    c = np.abs(np.sin(a*0.5))


    line.set_color((c,0,1.0))
   # line.set_linewidth(5)
    #ax.view_init(elev=30, azim=frame)

    return line,

anim = FuncAnimation(fig, update, frames=300, interval=30, init_func=init, blit=True)

#anim.save("lissajous3d.gif", writer="pillow", fps=30)


plt.show()
```
