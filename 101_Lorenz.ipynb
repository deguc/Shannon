```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def func(t,var,*args):

    x,y,z = var
    s,r,b = args

    dx = s*(y-x)
    dy = x*(r-z) - y
    dz = x*y - b*z

    return [dx,dy,dz]

t_span = (0,100)
t_eval = np.linspace(0,30,30000)
init = (0.1,0.1,0.1)
s,r,b = 10,28,8/3

sol = solve_ivp(func,t_span,init,args=(s,r,b),t_eval=t_eval,method='LSODA')
x,y,z = sol.y

plt.subplot(projection='3d')
plt.title('Lorenz Attractor')
plt.plot(x,y,z)
plt.show()
```
